# Manejo de vecotes con matplotlib

```python
import numpy as np
import numpy.linalg as npl
import matplotlib.pyplot as plt 
```

```python
Vector_1 = np.array([1, 3])  # x,y,z
Vector_2= np.array([2, 2])
```

```python
#funcion donde solo agarrara en x e y
def graficarVectores(vecs,cols,limit=([0,4],[0,4]),nombre='grafica'):
  plt.figure(figsize=(5,5))
  nombres=[]
  for i in range(len(vecs)):
    vec=vecs[i]
    plt.quiver(0,0,vec[0],vec[1],color=cols[i],angles='xy',scale_units='xy',scale=1)
    nombres.append(f'Vector {i}')
  plt.legend(nombres)
  plt.xlim(limit[0][0], limit[0][1])
  plt.ylim(limit[1][0],limit[1][1])
  plt.title(nombre)
  plt.grid()
```

```python
#funcion donde agarrara en x, y, z 
def graficarVectores_3(vecs,cols,angulos=[20,-15],Nombre='Grafica'):
  fig = plt.figure()
  ax = plt.axes(projection ='3d')
  for i in range(len(vecs)):
    vec=vecs[i]
    ax.plot3D([0,vec[0]],[0,vec[1]],[0,vec[2]],color=cols[i],label=f'vector {i}')
    ax.view_init(angulos[0], angulos[1]) 
  ax.set_xlabel('Eje x')
  ax.set_ylabel('Eje y')
  ax.legend()
  ax.set_title(Nombre)
  plt.grid()
```

```python
graficarVectores([Vector_1,Vector_2],['blue','red'],([0,2.5],[0,3.5]),'matriz1')
```

```python
#ejemplo de 3D
graficarVectores_3([[1,3,2],[2,4,1]],['blue','red'],[20, -50],'Grafica 3D')
```

*En la siguiente parte se buscara lo siguiente*

- Genera dos vectores en   y represntalos en un plano utilizando la librera Matplotlib.
- Calcula el producto punto entre los vectores generados.
- Calcula la longitud de ambos vectores.
- Encuentra el ngulo que forman los vectores entre s y genera un nuevo vector que sea ortogonal a cualquiera de los dos primeros.


https://numpy.org/doc/stable/reference/generated/numpy.linalg.norm.html
https://numpy.org/doc/stable/reference/generated/numpy.arccos.html

```python
#Se sacara el producto punto
Producto_punto=np.dot(Vector_1,Vector_2)
Longitud_1_vector=np.linalg.norm(Vector_1)
Longitud_2_vector=np.linalg.norm(Vector_2)
Angulo=np.arccos(np.dot(Vector_1,Vector_2)/(np.linalg.norm(Vector_1)*np.linalg.norm(Vector_2)))

print(f' la longitud del primer vector es {Longitud_1_vector} la longitud del segundo vector es {Longitud_2_vector}')
print(f' El producto punto es de {Producto_punto} ')
print(f' El Angulo es de {Angulo} ')
```

*Ejemplo*

> Sea el vector base $ b = \begin{bmatrix} 5 \\ 3 \end{bmatrix}\in \R^2$  que genera el subespacio unidimensional $ u\in \R^2$. encuentra el valor la cordenada () , la proyeccion $m_u(x)\in u$ y la matriz $P_x$ para proyectar sobre el subespacio $u$ el vector $ x = \begin{bmatrix} -2 \\ 7 \end{bmatrix}\in \R^2$

```python
b = np.vstack([5,3])
x = np.vstack([-2,7])

lambd = np.dot(b.T,x)/npl.norm(b)**2 # b^T x / ||b||^2;
vector_proyectado = b*lambd
matriz_proyeccion = np.dot(b,b.T)/npl.norm(b)**2 #bb^T/||b||^2

print(f' El valor cordenada es {lambd} ')
print(f' La proyeccion {vector_proyectado[0]} y {vector_proyectado[1]}  ')
print(f' La Matriz para proyectar sobre el subespacio {matriz_proyeccion[0]} y {matriz_proyeccion[1]}')
```

```python
graficarVectores([b,vector_proyectado,x],['Black','blue','red'],([-4,6],[-1,8]),'Pronostico')
```

https://numpy.org/doc/stable/reference/generated/numpy.trace.html
https://numpy.org/doc/stable/reference/generated/numpy.poly.html
https://numpy.org/doc/stable/reference/generated/numpy.roots.html

```python
Matriz = np.array([[6,2,5,7],[3,6,1,2],[5,8,4,3],[1,2,1,4]])

Determinante = np.linalg.det(Matriz)
Taza=np.trace(Matriz)  #taza
Rango=npl.matrix_rank(Matriz) #rango
coeficiente = np.poly(Matriz)  #coeficientes de su polinomio
Raiz = np.roots(coeficiente)  #raiz

print(f' la determinante de la matriz es {Determinante}')
print(f' La taza de la matriz {Taza} ')
print(f' El rango de la matriz es {Rango} ')
print(f' El coeficiente de la matriz es {coeficiente}\n ')
print(f' la Raiz de la matriz es {Raiz} ')
```
