```python
import numpy as np
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
```

```python
dataset = load_iris()
x=pd.DataFrame(dataset.data,columns=dataset.feature_names)
y=pd.DataFrame(dataset.target,columns=['target'])
```

```python
# Dividiendo eI conjunto de datos en conauntos de entrenamiento y de prueba
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2,random_state=42)
# Realizando e I agrupamiento utilizando el algoritmo k—means
kmeans = KMeans(n_clusters=3,random_state=42)
kmeans.fit(x_train)
```

```python
kmeans.cluster_centers_[:,0]
```

```python
plt.scatter (x_train.iloc[:,0] , x_train.iloc[:,1], c=kmeans.labels_ ,cmap="YlGnBu")
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='o', s=300, c='red')
plt.title(' Agrupamiento k—means con iris_dataset')
plt.xlabel( 'Longitud del sépalo (cm) ' )
plt.ylabel( 'Anchura del sépalo (cm) ' )
plt.show ( )
```
