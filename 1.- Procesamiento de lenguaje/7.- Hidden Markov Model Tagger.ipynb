 **5.	Implementa una función que calcule la distribución de frecuencias de las etiquetas en una categoría dentro del corpus Brown y con base en un esquema de etiquetado. Usa el esqueleto propuesto en la figura 14.**

```python
# Se define la función de explora_tagset_fd para trabajar con los parametros en el etiqutado del corpus.
def explora_tagset_fd(genero, tagset):
    palabras_universal = brown.tagged_words(categories=genero, tagset=tagset)
    tags = [w for w in palabras_universal]
    tags = nltk.FreqDist(tag for (word, tag) in palabras_universal)
    etiquetas = tags
    num_etiquetas = len(etiquetas)
    top_tags = tags.most_common()
    return num_etiquetas, top_tags
```

 **7.	Entrena y evalúa un etiquetador HMM con datos de entrenamiento y prueba respectivamente. Utiliza el esqueleto de la función propuesta en la figura 15**


```python
# Se declara la funcion de entrenamientoHMM para el uso del entrenamiento del cropus trabajado.
def entrenamientoHMM(oracion, tamanio):
    oraciones_etiq = brown.tagged_sents(categories='news')
    # Datos de entrenamiento.
    datos_train =oraciones_etiq[-tamanio:]
    # Datos de prueba.
    datos_test = oraciones_etiq[:100]
    #se entrena el modelo
    trainer = hmm.HiddenMarkovModelTrainer()
    # Entrena el modelo HiddenMArkovModelTagger con el método train().
    tagger = trainer.train_supervised(datos_train)
    # Utiliza el etiquetador para etiquetar la oración.
    oracion_etiqueta_hmm = tagger.tag(oracion)
    # Uilizando el etiquetador, evalua la precisión (accuracy) con los datos de prueba.
    acc = tagger.evaluate(datos_train)
    return tagger, oracion_etiqueta_hmm, acc
```
