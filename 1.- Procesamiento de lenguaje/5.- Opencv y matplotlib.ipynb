```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def Multiples_imagenes(imagen,titulos):#incompleta
  cantidad=int(len(imagen))
  if cantidad==1584:
    plt.title(titulos)
    plt.imshow(imagen)
    plt.show
  else:
    if cantidad%2==0:
      for i in range(cantidad):
        plt.subplot(2,(int(cantidad/2)),i+1),plt.imshow(imagen[i],vmin=0,vmax=255) #subplot(cantidad_y,cantidad_x,crecimiento)
        plt.title(titulos[i])
        plt.xticks([]),plt.yticks([])
      plt.show()
    else:
      for i in range(cantidad):
        plt.subplot(2,(int(cantidad/2)+1),i+1),plt.imshow(imagen[i],vmin=0,vmax=255) #subplot(cantidad_y,cantidad_x,crecimiento)
        plt.title(titulos[i])
        plt.xticks([]),plt.yticks([])
      plt.show()
```

**8.	Ejecuta un desplazamiento de 50 pixeles sobre el eje horizontal de las imágenes originales.**

```python
alto_1,ancho_1,largo1=img1_cv2_rgb.shape
alto_2,ancho_2,largo2=img2_cv2_rgb.shape
```

```python
M1 = np.float32([[1,0,50],[0,1,0]])    # y aplicando la función de warAffine para realizar el desplazamiento de la imagen 1.
image1_pix = cv.warpAffine(img1_cv2_rgb,M1,(ancho_1,alto_1))
image2_pix = cv.warpAffine(img2_cv2_rgb,M1,(ancho_2,alto_2))
```

```python
fig, axs= plt.subplots(2,1,figsize=(15, 5))
axs[0].imshow(image1_pix)
axs[1].imshow(image2_pix)
plt.show()
```

 **9.	Realiza una rotación de 45 grados de las imágenes originales.**

```python
MR1 = cv.getRotationMatrix2D((ancho_1//2,alto_1//2),45,1)
image_1 = cv.warpAffine(img1_cv2_rgb,MR1,(ancho_1,alto_1))

MR2 = cv.getRotationMatrix2D((ancho_2//2,alto_2//2),45,1)
image_2 = cv.warpAffine(img2_cv2_rgb,MR1,(ancho_2,alto_2))

fig, axs= plt.subplots(2,1,figsize=(15, 5))
axs[0].imshow(image_1)
axs[1].imshow(image_2)
plt.show()
```
